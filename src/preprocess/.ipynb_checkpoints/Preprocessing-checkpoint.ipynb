```python
import numpy as np
import pandas as pd
import os
```

```python
main_dir = os.getcwd()
```

## Function definition

```python
def merge_df(file_list):
    counter = 1
    df_list = pd.DataFrame()
    for file in file_list:
        if (os.stat(file).st_size != 0):
            df = pd.read_csv(file, sep = ",")
            df['symbol'] = file
            df_list = df_list.append(df)
            print (counter, " out of ", len(file_list))
            counter += 1
    return pd.DataFrame(df_list)
```

## Concatenating stocks data

```python
# STOCKS
os.chdir(main_dir)
os.chdir("stocks/Data/Stocks")
stock_list = os.listdir()
```

```python
stocks = merge_df(stock_list)
```

## Concatenating ETFs data

```python
# ETFs
os.chdir(main_dir)
os.chdir("stocks/Data/ETFs")
etf_list = os.listdir()
```

```python
etf = merge_df(etf_list)
```
